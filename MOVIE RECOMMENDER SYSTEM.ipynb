```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=500, stop_words='english') #menghapus kata yang sama seperti = the, a, has, have
```

```python
vectors = cv.fit_transform(mov_df['tags']).toarray() 
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
 
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key=lambda x:x[1])[1:6]
```

```python
def recommend(mov):
    mov_index = mov_df[mov_df['title'] == mov].index[0]
    distances = similarity[mov_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(mov_df.iloc[i[0]].title) 
```

```python
recommend('Dear Nathan')
```
